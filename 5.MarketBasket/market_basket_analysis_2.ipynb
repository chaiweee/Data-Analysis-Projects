```python
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df = pd.read_csv('market_basket_dataset_2.csv', sep=';', engine='python')
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
new_df = df[df['Itemname'].notna()]
new_df.info()
```

```python
print(new_df.isnull().sum())
```

### Check if there are any pattern for country

```python
#check if there are any patterns for country and what is the share of each country
new_df.groupby(['Country'])['BillNo'].nunique().sort_values(ascending=False)
```

### Prepare the data for association rule

```python
dfprep = new_df.groupby(['BillNo','Itemname']).agg({'Quantity':'sum'}).reset_index()
dfprep
```

```python
dfprep = new_df.groupby(['BillNo','Itemname']).agg({'Quantity':'sum'}).reset_index().pivot(index='BillNo',columns='Itemname').fillna(0)
dfprep
```

```python
dfprep.columns = dfprep.columns.droplevel(0)
```

```python
dfprep.head()
```

```python
dfprep.reset_index(inplace=True)
dfprep
```

```python
#total number of items
dfprep.drop(dfprep.columns[0],inplace=True,axis=1)
dfprep
```

```python
dfprep = dfprep.applymap(lambda x: True if x>0 else False)
dfprep
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(dfprep, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.5)
rules.shape
```
